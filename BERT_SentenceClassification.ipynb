## Data processing

```python
import pandas as pd
import datasets
from datasets import Dataset, DatasetDict
 
# df = pd.read_csv(base_dir + 'sentences_classification.csv')
df = pd.read_csv('NLP_Project_dataset2.csv')

df = df.filter(['sentence','label'], axis=1)
df.head()

```

```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(df, test_size=0.2,random_state=99, stratify = df['label'])
df_train, df_val = train_test_split(df_train, test_size=0.25,random_state=99, stratify = df_train['label'])
print(df_train.shape)
print(df_val.shape)
print(df_test.shape)
```

```python
df_train.head()
```

```python
df_val.head()
```

```python
df_test.head()
```

```python
train = Dataset.from_pandas(df_train)
val = Dataset.from_pandas(df_val)
test = Dataset.from_pandas(df_test)
 
 
dataset = DatasetDict()
 
dataset['train'] = train
dataset['val'] = val
dataset['test'] = test
 
dataset
```

```python
dataset['train'][0]
```

##Bert base multilingual

### Tokenizer

```python
from transformers import  AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-multilingual-cased")
print(tokenizer.decode(tokenizer('ทดสอบ')["input_ids"]))

def preprocess_function(examples):
    return tokenizer(examples["sentence"], truncation=True)
  
```

```python
tokenized_dataset = dataset.map(preprocess_function, batched=True)
tokenized_dataset['train'].shape
```

```python
from transformers import DataCollatorWithPadding

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
import evaluate

accuracy = evaluate.load("accuracy")
import numpy as np

def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=1)
    return accuracy.compute(predictions=predictions, references=labels)

id2label = {0:"Request", 1: "Neutral"}
label2id = {"Request":0  ,"Neutral": 1}
```

```python
from transformers import create_optimizer
import tensorflow as tf

batch_size = 16
num_epochs = 5
batches_per_epoch = len(tokenized_dataset["train"]) // batch_size
total_train_steps = int(batches_per_epoch * num_epochs)
optimizer, schedule = create_optimizer(init_lr=2e-5, num_warmup_steps=0, num_train_steps=total_train_steps)
```

```python
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer

model = AutoModelForSequenceClassification.from_pretrained("bert-base-multilingual-cased", num_labels=2, id2label=id2label, label2id=label2id)
```

```python
training_args = TrainingArguments(
    output_dir="laptop_sentence_classfication_BERT",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=20,
    weight_decay=0.01,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    push_to_hub=False,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset= tokenized_dataset["train"],
    eval_dataset= tokenized_dataset["val"],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)

trainer.train()
```

```python
# trainer.push_to_hub('TirkNork/laptop_sentence_classfication_BERT')
```

```python
def metric_cal(p,l):

  TP_request = 0
  TN_request = 0
  FP_request = 0
  FN_request = 0

  TP_neutral = 0
  TN_neutral = 0
  FP_neutral = 0
  FN_neutral = 0

  count = 0

  for (i,j) in zip(predictions, labels):
    if i == 0 and j == 0:
      TP_request += 1
      TN_neutral += 1
      count += 1
    elif i ==1 and j == 0:
      FN_request += 1
      FP_neutral += 1
    elif i == 0 and j == 1:
      FN_neutral += 1
      FP_request += 1
    elif i == 1 and j == 1:
      TP_neutral += 1
      TN_request += 1
      count += 1

  pre_request = TP_request/(TP_request+FP_request)
  recall_request = TP_request/(TP_request+FN_request)
  f1_request = 2*pre_request*recall_request/(pre_request+recall_request)

  pre_neutral = TP_neutral/(TP_neutral+FP_neutral)
  recall_neutral = TP_neutral/(TP_neutral+FN_neutral)
  f1_neutral = 2*pre_neutral*recall_neutral/(pre_neutral+recall_neutral)

  print("Acc :",count/len(predictions))
  print("Precision request:",pre_request)
  print("Recall request:",recall_request)
  print("F1 request:",f1_request)
  print("Precision neutral:",pre_neutral)
  print("Recall neutral:",recall_neutral)
  print("F1 neutral:",f1_neutral)
```

```python
#train
predictions, labels, _ = trainer.predict(tokenized_dataset["val"])
predictions = np.argmax(predictions, axis=-1)

metric_cal(predictions, labels)
```

```python
#test
predictions, labels, _ = trainer.predict(tokenized_dataset["test"])
predictions = np.argmax(predictions, axis=-1)

metric_cal(predictions, labels)
```
